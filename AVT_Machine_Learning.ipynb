```python
import pandas as pd

# Use the following lines to upload the CSV file from your local system
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a DataFrame
file_path = '/content/AVT ML_Work.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
def calculate_values(fresh_marigold_kg):
    # Define ratios
    hexane_ratio = 3
    methanol_ratio = 3

    # Calculate values based on the given ratios
    first_hexane_wash_hexane_kg = fresh_marigold_kg * hexane_ratio
    first_hexane_wash_miscella_out_kg = first_hexane_wash_hexane_kg * 0.07
    second_hexane_wash_hexane_kg = first_hexane_wash_hexane_kg
    second_hexane_wash_miscella_out_kg = first_hexane_wash_hexane_kg * 0.05
    second_part_of_second_hexane_wash_kg = first_hexane_wash_hexane_kg * 0.045
    second_part_of_second_hexane_wash_miscella_out_kg = second_part_of_second_hexane_wash_kg * 0.03
    third_hexane_wash_hexane_kg = first_hexane_wash_hexane_kg
    third_hexane_wash_miscella_out_kg = first_hexane_wash_hexane_kg * 0.06
    second_part_of_third_hexane_wash_kg = first_hexane_wash_hexane_kg * 0.035
    second_part_of_third_hexane_wash_miscella_out_kg = second_part_of_third_hexane_wash_kg * 0.02
    third_part_of_third_hexane_wash_kg = first_hexane_wash_hexane_kg * 0.025
    third_part_of_third_hexane_wash_miscella_out_kg = third_part_of_third_hexane_wash_kg * 0.015
    fourth_hexane_wash_kg = first_hexane_wash_hexane_kg * 0.01
    fourth_hexane_wash_miscella_out_kg = fourth_hexane_wash_kg * 0.005

    total_miscella_in_combined_hep_kg = first_hexane_wash_miscella_out_kg + second_hexane_wash_miscella_out_kg + \
                                        second_part_of_second_hexane_wash_miscella_out_kg + third_hexane_wash_miscella_out_kg + \
                                        second_part_of_third_hexane_wash_miscella_out_kg + third_part_of_third_hexane_wash_miscella_out_kg + \
                                        fourth_hexane_wash_miscella_out_kg

    first_methanol_wash_methanol_kg = fresh_marigold_kg * methanol_ratio
    first_methanol_wash_miscella_out_kg = first_methanol_wash_methanol_kg * 0.07
    second_methanol_wash_methanol_kg = first_methanol_wash_methanol_kg
    second_methanol_wash_miscella_out_kg = first_methanol_wash_methanol_kg * 0.05
    second_part_of_second_methanol_wash_kg = first_methanol_wash_methanol_kg * 0.045
    second_part_of_second_methanol_wash_miscella_out_kg = second_part_of_second_methanol_wash_kg * 0.03
    third_methanol_wash_methanol_kg = first_methanol_wash_methanol_kg
    third_methanol_wash_miscella_out_kg = first_methanol_wash_methanol_kg * 0.06
    second_part_of_third_methanol_wash_kg = first_methanol_wash_methanol_kg * 0.035
    second_part_of_third_methanol_wash_miscella_out_kg = second_part_of_third_methanol_wash_kg * 0.02
    third_part_of_third_methanol_wash_kg = first_methanol_wash_methanol_kg * 0.025
    third_part_of_third_methanol_wash_miscella_out_kg = third_part_of_third_methanol_wash_kg * 0.015
    fourth_methanol_wash_kg = first_methanol_wash_methanol_kg * 0.01
    fourth_methanol_wash_miscella_out_kg = fourth_methanol_wash_kg * 0.005

    total_miscella_in_combined_mep_kg = first_methanol_wash_miscella_out_kg + second_methanol_wash_miscella_out_kg + \
                                        second_part_of_second_methanol_wash_miscella_out_kg + third_methanol_wash_miscella_out_kg + \
                                        second_part_of_third_methanol_wash_miscella_out_kg + third_part_of_third_methanol_wash_miscella_out_kg + \
                                        fourth_methanol_wash_miscella_out_kg

    amount_of_acetone_added_kg = fresh_marigold_kg * 0.1
    miscella_weight_after_stripping_mg_wet_cake_kg = total_miscella_in_combined_hep_kg + total_miscella_in_combined_mep_kg - amount_of_acetone_added_kg
    weight_after_dhad_kg = miscella_weight_after_stripping_mg_wet_cake_kg * 0.9
    total_carotenoids = weight_after_dhad_kg * 0.1
    total_residual_solvent = weight_after_dhad_kg * 0.05

    return {
        'First Hexane Wash Hexane (kg)': first_hexane_wash_hexane_kg,
        'First Hexane Wash Miscella OUT (kg)': first_hexane_wash_miscella_out_kg,
        'Second Hexane Wash Hexane (kg)': second_hexane_wash_hexane_kg,
        'Second Hexane Wash Miscella OUT (kg)': second_hexane_wash_miscella_out_kg,
        'Second Part of Second Hexane Wash (kg)': second_part_of_second_hexane_wash_kg,
        'Second Part of Second Hexane Wash Miscella OUT (kg)': second_part_of_second_hexane_wash_miscella_out_kg,
        'Third Hexane Wash Hexane (kg)': third_hexane_wash_hexane_kg,
        'Third Hexane Wash Miscella OUT (kg)': third_hexane_wash_miscella_out_kg,
        'Second Part of Third Hexane Wash (kg)': second_part_of_third_hexane_wash_kg,
        'Second Part of Third Hexane Wash Miscella OUT (kg)': second_part_of_third_hexane_wash_miscella_out_kg,
        'Third Part of Third Hexane Wash (kg)': third_part_of_third_hexane_wash_kg,
        'Third Part of Third Hexane Wash Miscella OUT (kg)': third_part_of_third_hexane_wash_miscella_out_kg,
        'Fourth Hexane Wash (kg)': fourth_hexane_wash_kg,
        'Fourth Hexane Wash Miscella OUT (kg)': fourth_hexane_wash_miscella_out_kg,
        'Total Miscella IN Combined HEP (kg)': total_miscella_in_combined_hep_kg,
        'First Methanol Wash Methanol (kg)': first_methanol_wash_methanol_kg,
        'First Methanol Wash Miscella OUT (kg)': first_methanol_wash_miscella_out_kg,
        'Second Methanol Wash Methanol (kg)': second_methanol_wash_methanol_kg,
        'Second Methanol Wash Miscella OUT (kg)': second_methanol_wash_miscella_out_kg,
        'Second Part of Second Methanol Wash (kg)': second_part_of_second_methanol_wash_kg,
        'Second Part of Second Methanol Wash Miscella OUT (kg)': second_part_of_second_methanol_wash_miscella_out_kg,
        'Third Methanol Wash Methanol (kg)': third_methanol_wash_methanol_kg,
        'Third Methanol Wash Miscella OUT (kg)': third_methanol_wash_miscella_out_kg,
        'Second Part of Third Methanol Wash (kg)': second_part_of_third_methanol_wash_kg,
        'Second Part of Third Methanol Wash Miscella OUT (kg)': second_part_of_third_methanol_wash_miscella_out_kg,
        'Third Part of Third Methanol Wash (kg)': third_part_of_third_methanol_wash_kg,
        'Third Part of Third Methanol Wash Miscella OUT (kg)': third_part_of_third_methanol_wash_miscella_out_kg,
        'Fourth Methanol Wash (kg)': fourth_methanol_wash_kg,
        'Fourth Methanol Wash Miscella OUT (kg)': fourth_methanol_wash_miscella_out_kg,
        'Total Miscella IN Combined MEP (kg)': total_miscella_in_combined_mep_kg,
        'Amount of Acetone Added (kg)': amount_of_acetone_added_kg,
        'Miscella Weight After Stripping MG Wet Cake (kg)': miscella_weight_after_stripping_mg_wet_cake_kg,
        'Weight After DHAD (kg)': weight_after_dhad_kg,
        'Total Carotenoids (kg)': total_carotenoids,
        'Total Residual Solvent (kg)': total_residual_solvent
    }

# Input
```

```python
# Input weight of Fresh Marigold Raw Material IN (kg)
fresh_marigold_kg = 75500

# Calculate the values
calculated_values = calculate_values(fresh_marigold_kg)

# Print the output values
for key, value in calculated_values.items():
    print(f"{key}: {value}")
```
